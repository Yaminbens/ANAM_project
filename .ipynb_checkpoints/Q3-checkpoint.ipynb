3.1 - load data

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# import plotly.plotly as py
%matplotlib inline

from sklearn.datasets import fetch_lfw_people
lfw_people = fetch_lfw_people()
```

```python
from random import sample
from sklearn.feature_extraction import image
images = lfw_people.images[:50]

#normalize pixels
for im in images:
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            im[i,j] = im[i,j]/255

```

```python
#create training data
patches_list = []
training_data = []
prow,pcol = 6,6
for im in images:
    patches = image.extract_patches_2d(im,patch_size=(prow,pcol), max_patches=60)
    patches_list.append(patches)
    indices = sample(xrange(60), 30)
    training_data_tmp = [patches[i] for i in indices]
    training_data.extend(training_data_tmp)

training_data = np.asarray(training_data)
```

3.2 PCA on training data

```python
from sklearn import decomposition
# from sklearn.decomposition import RandomizedPCA

training_data_reshaped = []
training_data_reduced = []
nn_components = 36

#reshape data to 36 vec
for patch in training_data:
    training_data_reshaped.append(np.reshape(patch,36))

training_data_reshaped = np.asarray(training_data_reshaped)


pca = decomposition.PCA(n_components=nn_components, svd_solver='auto') # creating a PCA object
pca.fit(training_data_reshaped)
training_data_reduced = pca.transform(training_data_reshaped)
reconstruct = training_data_reduced.dot(pca.components_)
```

```python
xim = np.reshape(reconstruct[0],(prow,pcol))
fig = plt.figure()
a=fig.add_subplot(1,2,1)
imgplot = plt.imshow(training_data[0])
a.set_title('Before')
imgplot.set_cmap('nipy_spectral')
plt.colorbar()

a=fig.add_subplot(1,2,2)
imgplot = plt.imshow(xim)
a.set_title('After')
imgplot.set_cmap('nipy_spectral')
plt.colorbar()

```

```python
n_components = 25
h, w = training_data[0].shape

pca = decomposition.PCA(n_components=n_components, whiten=True, svd_solver='randomized').fit(training_data_reshaped) 
re_images = pca.components_.reshape((n_components, h, w))
re_images.shape
```
