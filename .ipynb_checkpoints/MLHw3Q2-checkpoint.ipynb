```python
# From calculation, it is expected that the local minimum occurs at x=9/4
import math
from copy import copy

steps = []

cur_x = [1.0,1.0] # The algorithm starts at x=6
ni = 0.001 # step size multiplier
l = 0.99
m =[0,0]
precision = 0.0001
previous_step_size = 100

# f =80*x**4 +0.01*y**6

dfx = lambda x: 4*80*x**3 
dfy = lambda y: 0.01*6*y**5

while previous_step_size > precision:
    prev_x = copy(cur_x)
    prev_m = copy(m)
    steps.append(copy(prev_x))
    m[0] = l*prev_m[0] + ni*dfx(prev_x[0])
    m[1] = l*prev_m[1] + ni*dfy(prev_x[1])
    cur_x[0] += -m[0]
    cur_x[1] += -m[1]
#     print((cur_x[0] - prev_x[0])**2)
    previous_step_size = math.sqrt(((cur_x[0] - prev_x[0])**2)+((cur_x[1] - prev_x[1])**2))
#     print("p:",previous_step_size)

print("The local minimum occurs at {},{}".format(cur_x[0],cur_x[1]))
print("number of steps: {}".format(len(steps)))
steps = np.asarray(steps)
```
